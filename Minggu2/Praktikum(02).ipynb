  - Pengguna dapat mengklik, zoom, menyimpan gambar, dan menambahkan anotasi

```python
import random

# Buat gambar berukuran 400x400 piksel dengan latar belakang putih
gambar = np.zeros((400, 400, 3), dtype=np.uint8) * 255

# Buat 5 kotak dengan ukuran, posisi, dan warna acak
for _ in range(5):
    ukuran = random.randint(20, 100)
    x = random.randint(0, 300)
    y = random.randint(0, 300)
    warna = (random.randint(0, 255), random.randint(0, 255), random.randint(0, 255))

    cv.rectangle(gambar, (x, y), (x + ukuran, y + ukuran), warna, -1)

# Tampilkan gambar
cv2_imshow(gambar)
cv.waitKey(0)
cv.destroyAllWindows()
```

Buat rectangle pada bagian sudut bawah kiri channel B pada color space RGB
dari citra kitten/ lena/ mandrill/ male/ female/ couple/ sailboat/ peppers!

```python
path ='/content/drive/MyDrive/Colab Notebooks/pemandangan.png'
pemandangan = cv.imread(path)

blue_channel, green_channel, red_channel = cv.split(pemandangan)
height, width, _ = pemandangan.shape
cv.rectangle(blue_channel, (0, 120), (0 + 120, (height - 200) + 200), (255,255,255), -1)
modified_pemandangan = cv.merge((blue_channel, green_channel, red_channel))
cv2_imshow(modified_pemandangan)
```

```python
cv.putText(modified_pemandangan, 'Pemandangan', (100, 200), cv.FONT_ITALIC, 1, (255, 255, 255), thickness=2)
cv2_imshow(modified_pemandangan)
```
